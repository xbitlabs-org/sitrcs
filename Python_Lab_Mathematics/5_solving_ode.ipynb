```python
# Solve dy/dx + ytanx = secx
```

```python
from sympy import Function, dsolve, Eq, Derivative, sin, cos, tan, sec, symbols

# Define the variables
x = symbols('x')
print('x is', x)
y = Function('y')(x)
print('y is :', y)
# Define the differential equation
ode = Eq(Derivative(y, x) + y * tan(x), sec(x))

# Solve the ODE
solution = dsolve(ode)

# Display the solution
print("Solution:")
print(solution)
```

```python
# Solve  :  dp(t) / dt = r
```

```python
# Method 1
from sympy import symbols, Function, Eq, dsolve

# Define the variables
t, r = symbols('t r')
p = Function('p')(t)

# Define the differential equation
ode = Eq(p.diff(t), r)

# Solve the ODE
solution = dsolve(ode)

# Display the solution
print("Solution:")
print(solution)
```

```python
# Solve dy/dt = -ky   with k = 0.3 and y(0) =5
```

```python
from sympy import symbols, Function, Eq, dsolve

# Define the variable and constant
t, k = symbols('t k')
y = Function('y')(t)

# Define the differential equation
ode = Eq(y.diff(t), -k * y)

# Define the initial condition
initial_condition = {y.subs(t, 0): 5}

# Solve the ODE with the initial condition
solution = dsolve(ode, ics=initial_condition)

# Display the solution
print("Solution:")
print(solution)
```
