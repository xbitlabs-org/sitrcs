```python
from sympy import plot_implicit, symbols, Eq
```

```python
# Q1 ) Plot x^2 + y^2 = 5

x, y = symbols('x y')
equation = Eq(x**2 + y**2, 5)
plot_implicit(equation)
```

```python
plot_implicit(equation, (x, -10, 10), (y,-41, 4))
```

```python
# Q2 ) Plot y^2 ( a-x )  = x^2 ( a + x )  where a > 0
```

```python
from sympy import symbols, Eq, plot_implicit

# Define symbols
a, x, y = symbols('a x y')

a = 2
# Define the equation
equation = Eq(y**2 * (a - x), x**2 * (a + x))

# Specify the range for x and y
#range_x = (-5, 5)
#range_y = (-5, 5)

# Plot the implicit equation
#plot_implicit(equation, (x, *range_x), (y, *range_y), title='Plot of y^2(a-x) = x^2(a+x)')

plot_implicit(equation)
```
