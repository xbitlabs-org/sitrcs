```python
# 2-D Plots of cartesian and polar curve
import math
import matplotlib.pyplot as plt 
```

```python
# PLotting Exponential plots
x = []
y = []
```

```python
i = -10
while i <= 10:
    i = i+0.01
    #print(i)
    x.append(i)
    j = math.exp(i)
    y.append(j)
```

```python
plt.plot(x,y)
```

```python
sinx = []
cosx = []
```

```python
for i in x:
    j = math.sin(i)
    sinx.append(j)
    k = math.cos(i)
    cosx.append(k)
```

```python
plt.plot(x, sinx)
plt.plot(x, cosx)
```
