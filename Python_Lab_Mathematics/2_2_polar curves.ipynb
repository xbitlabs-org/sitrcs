```python
# Q2 Cardioid - specific shape that can be represented in polar coordinates
# r = 5(1+cosx) where x = theta ( angle ) 
```

```python
theta = np.linspace(0, 2*np.pi, 1000)  # The function np.linspace(0, 2*np.pi, 1000) generates an array of 1000 equally spaced values between 0 and 2pi ( inclusive ) 
```

```python
r = 5 + 5 * cos(theta)
```

```python
plt.polar(theta, r, 'r')
```

```python
# Q3 PLot r = 2|cos2x|   ( 4 leaved rose ) 

theta = np.linspace(0, 2*pi, 50)
```

```python
r = 2*abs(cos(2*theta))
```

```python
plt.polar(theta, r, 'r')
```

```python
# Q5 cardioids r = a + acos(x)    and    r = a - acos(x)   where x = theta ( angle ) 
```

```python
plt.axes(projection='polar')
a = 3

rad = np.arange(0, 2*np.pi, 0.01)

for i in rad:
    r = a + ( a*np.cos(i))
    plt.polar(i, r, 'g')
    r2 = a - ( a*np.cos(i))
    plt.polar(i, r2, 'r')
```
