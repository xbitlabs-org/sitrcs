```python
import matplotlib.pyplot as plt
```

```python
# 1. Plotting points
x = [1,2,3,4,5,6,7,8]
y = [2,7,9,1,5,10,3,9]

#plt.scatter(x,y)
plt.plot(x,y, 'r+')
plt.plot(x,y, 'bo')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scattered points')
plt.show()
```

```python
# 2. Exponential curves
```

```python
import numpy as np

x = np.arange(start=-10, stop=10, step=0.001)

print(x)

y = np.exp(x)

print(y)

plt.plot(x,y)
```

```python
np.exp(-10)
```

```python
# Example power of a number raised to another power using arrays in numpy

base = np.array([2, 3, 4])

base

exponent = np.array([2, 3, 4])
exponent
```

```python
result = np.power(base, exponent)

result
```

```python
# Example : Sine and cosine curves 

x = np.arange(start=-10, stop=10, step=0.001)
x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)

#plt.title('x1 y1 x y2')
plt.title('Sine curve and cosine curve')
plt.xlabel('Value of x')
plt.ylabel('Value of y1 and y2')

plt.plot(x, y1, label='Sine curve')  # Plot the sine curve
plt.plot(x, y2, label='Cosine curve')  # Plot the cosine curve

plt.legend()

plt.grid()

plt.grid()
plt.show()
```
